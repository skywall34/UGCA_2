```python
# Dependencies
import tweepy           # To consume Twitter's API
import pandas as pd     # To handle data
import numpy as np      # For number computing

# For plotting and visualization:
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
import time
import os
%matplotlib inline
```

```python
from credentials import *
# importing from the credentials.py file

# API's setup:
def twitter_setup():
    """
    Utility function to setup the Twitter's API
    with our access keys provided.
    """
    # Authentication and access using keys:
    auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
    auth.set_access_token(OAUTH_TOKEN, OAUTH_TOKEN_SECRET)

    # Return API with authentication:
    api = tweepy.API(auth)
    return api
```

```python
import csv
import json

# We create an extractor object:
extractor = twitter_setup()

MAX_TWEETS = 50

# Mention the hashtag that you want to look out for

hashtag = "BETO"

tweets = tweepy.Cursor(extractor.search, q='BETO', tweet_mode="extended").items(MAX_TWEETS)

print ('Extracted ' + str(maximum_number_of_tweets_to_be_extracted) \
    + ' tweets with hashtag #' + hashtag)

    
```

```python
# Create JSON array
tweet_data = []

for tweet in tweets:
    tweet_data.append(json.loads(json.dumps(tweet._json)))

print (tweet_data[0])
```

```python
# Create the dataframe we will use
tweets_cleaned = pd.DataFrame()
# We want to know when a tweet was sent
tweets_cleaned['created_at'] = list(map(lambda tweet: time.strftime('%Y-%m-%d %H:%M:%S', time.strptime(tweet['created_at'],'%a %b %d %H:%M:%S +0000 %Y')), tweet_data))
# Who is the tweet owner
tweets_cleaned['user'] = list(map(lambda tweet: tweet['user']['screen_name'], tweet_data))
# How many follower this user has
tweets_cleaned['user_followers_count'] = list(map(lambda tweet: tweet['user']['followers_count'], tweet_data))
# What is the tweet's content
tweets_cleaned['full_text'] = list(map(lambda tweet: tweet['full_text'], tweet_data))
tweets_cleaned
```
