```python
import json
import pandas as pd
import time

tweet_json = open('fetched_tweets.json', 'r').read()
tweets = json.loads(tweet_json)
```

```python
for tweet in tweets:
    if 'retweeted_status' in tweet.keys():
        if(tweet['retweeted_status']['truncated'] == True):
            tweet['extended_tweet-full_text'] = tweet['retweeted_status']['extended_tweet']['full_text']
        
        else:
            tweet['extended_tweet-full_text'] = tweet['retweeted_status']['text']
        
    elif 'quoted_status' in tweet.keys():
        if(tweet['quoted_status']['truncated'] == True):
            tweet['extended_tweet-full_text'] = tweet['quoted_status']['extended_tweet']['full_text']
        else:
            tweet['extended_tweet-full_text'] = tweet['quoted_status']['text']
    elif 'extended_tweet' in tweet.keys():
        tweet['extended_tweet-full_text'] = tweet['extended_tweet']['full_text']
        
        
    else:
        tweet['extended_tweet-full_text'] = tweet['text']
    tweet['user-location'] = tweet['user']['location']
```

```python
for tweet in tweets:
    if 'retweeted_status' in tweet.keys():
        if tweet['retweeted_status']['user'] is not None:
            if tweet['retweeted_status']['user']['location'] is not None:
                tweet['location_retweet'] = tweet['retweeted_status']['user']['location']
            else:
                tweet['location_retweet'] = 'False'

    else:
        tweet['location_retweet'] = 'False'
```

```python
for tweet in tweets:
    if 'quoted_status' in tweet.keys():
        if tweet['quoted_status']['user'] is not None:
            if tweet['quoted_status']['user']['location'] is not None:
                tweet['quoted_location'] = tweet['quoted_status']['user']['location']

            else:
                tweet['quoted_location'] = 'False'
    else:
        tweet['quoted_location'] = 'False'
```

```python
tweets_cleaned = pd.DataFrame()
tweets_cleaned["location_retweet"] = list(map(lambda tweet: tweet["location_retweet"], tweets))
```

```python
# We want to know when a tweet was sent
tweets_cleaned['created_at'] = list(map(lambda tweet: time.strftime('%Y-%m-%d %H:%M:%S', time.strptime(tweet['created_at'],'%a %b %d %H:%M:%S +0000 %Y')), tweets))
# Who is the tweet owner
tweets_cleaned['user'] = list(map(lambda tweet: tweet['user']['screen_name'], tweets))
# How many follower this user has
tweets_cleaned['user_followers_count'] = list(map(lambda tweet: tweet['user']['followers_count'], tweets))
tweets_cleaned['user-location'] = list(map(lambda tweet: tweet['user-location'], tweets))
#tweets_cleaned['user-description'] = list(map(lambda tweet: tweet['user-description'], tweets))
# What is the tweet's content
tweets_cleaned['full_text'] = list(map(lambda tweet: tweet['extended_tweet-full_text'], tweets))
tweets_cleaned['location_retweet'] = list(map(lambda tweet: tweet['location_retweet'], tweets))
tweets_cleaned['quoted_location'] = list(map(lambda tweet: tweet['quoted_location'], tweets))
```

```python
# TODO , create a common column for the location
for tweet in tweets:
    # if both are false
    if tweet['location_retweet'] == 'False' and tweet['quoted_location'] == 'False':
        tweet['final_location'] = 'NaN'
       #print('Both are false')
    
    # if retweet is false
    elif tweet['location_retweet'] == 'False' and tweet['quoted_location'] != 'False':
        tweet['final_location'] = tweet['quoted_location']
        #print('Only retweet is false')
    
    # if quoted is false
    elif tweet['location_retweet'] != 'False' and tweet['quoted_location'] == 'False':
        tweet['final_location'] = tweet['location_retweet']
        #print('Only quoted is false')
    
    # if both are not false
        # if both are same
    else:
        if tweet['location_retweet'] == tweet['quoted_location']:
            tweet['final_location'] = tweet['location_retweet']
        else:
            new_string = tweet['location_retweet'] + ', ' + tweet['quoted_location']
            tweet['final_location'] = new_string

        #print('Both are not false')    
    
```

```python
location_dict = {'Houston, Texas' : 'bigcity ',
         'Dallas, TX' : 'bigcity ',
         'El Paso, TX' : 'bigcity ',
         'Austin, Texas' : 'bigcity ',
         'Austin, TX' : 'bigcity ',
         'Houston, TX' : 'bigcity ',
         'Houston' : 'bigcity ',
         'Houston Texas, USA' : 'bigcity ',
         'Austin/I-10/I-35' : 'bigcity ',
         'Washington, DC / San Antonio, TX' : 'bigcity ',
         'Fort Worth, TX' : 'bigcity ',
         'San Antonio, TX' : 'bigcity ',
         'Austin, Tx' : 'bigcity ',
         'Dallas TX' : 'bigcity ',
         'Austin TX' : 'bigcity ',
         'San Antonio, Texas' : 'bigcity ',
         'Houston, Texas, USA' : 'bigcity ',
         'El Paso, TX ;  Southern California' : 'bigcity ',
         'Austin' : 'bigcity ',
         'Dallas, Tx' : 'bigcity ',
         'Austin, Texas USA' : 'bigcity ',
         'Ft. Worth, Texas' : 'bigcity ',
         'ATX' : 'bigcity '
        }
import re
def multiple_replace(text):
    rx = re.compile('|'.join(map(re.escape, location_dict)))
    def swap_names(match):
        return location_dict[match.group(0)]
    return rx.sub(swap_names, text)
```

```python
def count_word(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

word_freq1 = count_word(str_clean1)
```

```python
def get_lift (name,issue,sortedLiftList):
    
    sum = 1841

    comb_denom = []
    for i in name:
        for j in issue:
            sub_l = [0,0]
            comb_denom.append(sub_l)
            sub_l[0] = (i[0],j[0])
            denom = (i[1]/sum) * (j[1]/sum)
            sub_l[1] = denom
    #print(comb_denom)
    comb_lift = []
    for i in comb_denom:
        for j in sortedLiftList:
            if i[0] == j[0]:
                sub_list = [0,0]
                comb_lift.append(sub_list)
                numer = j[1]/sum
                lift = numer/i[1]
                sub_list[0]=j[0]
                sub_list[1]=lift

    print(comb_lift)
    final_lift = []
    for i in comb_lift:
        value = i[1]
        final_lift.append(value)
    
    return final_lift
```
